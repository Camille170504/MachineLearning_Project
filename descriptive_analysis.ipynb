# Importation of the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Importation of the dataset

```python
#  Loading of the dataset from a CSV file
data=pd.read_csv("creditcard_2023.csv")

# Deleting "id" columns
data=data.drop(columns="id")

# Display the data to verify the changes
data
```

# Descriptive analysis of the data

## General analysis

```python
# Displaying the shape of the dataset (lines, columns)
print("Shape of the dataset : ", data.shape)

# Displays the types of variables
print("\nTypes of the variables : ", data.dtypes.unique())

# CHecking the missing values
print("\nMissing values : ", data.isnull().sum().sum())

# Brief statistical description of the numerical variables
print("\nBasic description of the dataset : ")
data.describe()
```

```python
print("\nGlobal distribution of the variable 'Class' (%):\n")

# Calculates the proportion of each class (fraud or not fraud)
print((data["Class"].value_counts(normalize=True)*100).round(3))
```

# Univariate analysis

```python
# Graphic style used
plt.style.use("seaborn-v0_8")

cols = data.columns
n = len(cols)

# Creating the graphs, 4 for each row
fig, axes = plt.subplots(-(-n // 4), 4, figsize=(20, 4 * (-(-n // 4))))

# Flattens the grid
axes = axes.flatten()

# Loop on all the columns to create the graphs
for i, col in enumerate(cols):
    sns.histplot(data[col], bins=30, kde=True, color='steelblue', edgecolor='black', ax=axes[i])
    axes[i].set_title(f"'{col}'", fontsize=12)
    axes[i].grid(True, linestyle='--', alpha=0.5)

# Deletes the unused axis to have a better display
for ax in axes[n:]:
    ax.remove()

# Automatic adjustment to avoid grahps' overlap
plt.tight_layout()
plt.show()
```

# Multivariate analysis

```python
# Figure's size
plt.figure(figsize=(12, 8))

# Creation of a heatmap based on the correlation matrix
sns.heatmap(data.corr(), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation matrix between numerical variables", fontsize=16)
plt.show()
```

```python
# Identifying the target variable for the correlation
corr_with_target = data.corr()['Class']

# Identifying the coluns we need to delete (columns with a correlation < 0.15 or > -0;15 with 'Class')
cols_to_drop = corr_with_target[(corr_with_target > -0.15) & (corr_with_target < 0.15)].index

# Keeping the column Amount for final analysis even if correlation is low
cols_to_drop = [col for col in cols_to_drop if col != 'Amount']

# Deleting the columns weekly correlated with 'Class'
df = data.drop(columns=cols_to_drop)

# Displaying de removed columns and the new dimensions of the dataset
print("Removed variables :", list(cols_to_drop))
print("New dimensions of the dataset :", df.shape)
```
